```python
questions = ["1. How many people have heart disease and how many doesn't have  heart disease ? ",
            "2. People of which sex has the most heart disease ? " ,
            "3. People of which sex has which type of chest pain most ? " ,
            "4. Peaple with which chest pain are most pron to have heart disease ? " ]

questions 
```

```python
# Let's find out the answer of the first question .

#  How many people have heart disease and how many doesn't have  heart disease ?

# getting the values
df.target.value_counts()
```

```python
# ploting bar chart
df.target.value_counts().plot(kind = 'bar' , color=['red' , 'blue'])
plt.title("Heart Disease Values")
plt.xlabel("1 = Heart Disease , 0 = No Heart Disease")
plt.ylabel("Amount")
```

```python
# plotting a pie chart
df.target.value_counts().plot(kind = 'pie' , figsize = (8 , 6))
plt.legend(["Disease" , "No Disease"]);
```

```python
# '0' represent 'Female'

# '1' represent 'Male'

# SEX column part

# '0' represent 'No Disease '

# '1' represent 'Disease '

#Target column part

# Now let's  check how many 'Male' and 'Female' are in the dataset

df.sex.value_counts()
```

```python
# plotting a pie chart
df.sex.value_counts().plot(kind = 'pie' , figsize = (8 , 6))
plt.legend(["Male" , "Female"]);
```

```python
# Let's find out the answer of the 2nd question .

# 2. People of which sex has the most heart disease  ?

pd.crosstab(df.target, df.sex)
```

```python
sns.countplot(x = 'target' , data = df, hue = 'sex')
plt.title("Heart Disease Frequency for Sex")
plt.xlabel("1 = Heart Disease , 0 = No Heart Disease")
```

```python
# Number of male is more than double in our dataset than female

# More than '45% male' has heart disease  and '75% female' has heart disease.
```

```python
# Let's find out the answer of the 3nd question .

# 3.People of which sex has which type of chest pain most ?   

df.cp.value_counts()
```

```python
# plotting a bar chart
df.cp.value_counts().plot(kind = 'bar' , color = ['red' , 'yellow' , 'blue' , 'khaki'])
plt.title('Chest pain type vs count');
```

```python
pd.crosstab(df.sex, df.cp)
```

```python
pd.crosstab(df.sex , df.cp).plot(kind = 'bar' , color= ['red' , 'yellow' , 'blue' , 'khaki'])
plt.title("Types of chest pain for sex")
plt.xlabel('0 = Female , 1 = Male')
```

```python
# Most of 'male' has 'type 0' chest pain and Least of 'male has 'type 4' pain

# in case of 'Female' 'type 0' and 'type 2' percentage is almost same 
```

```python
# Let's find out the answer of the 4rth  question .

# 4. Peaple with which chest pain are most pron to have heart disease ? 

pd.crosstab(df.cp, df.target)
```

```python
sns.countplot(x = 'cp' , data = df, hue = 'target');
```

```python
# Most of people who has 'type 0' chest pain has less chance of heart disease.

# And we see the opposite  for other types.
```
